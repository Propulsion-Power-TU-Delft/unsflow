```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import os
from Grid.src.functions import *
```

```python
with open('../SingleBlade/Output/NASAR37.pik', 'rb') as f:
    data = pickle.load(f)
```

```python
blade = data.blades[0]
kappa = blade.blade_metal_angle
```

```python
contour_template(blade.z_grid, blade.r_grid, 180/np.pi*kappa, r'$\kappa \quad \mathrm{[deg]}$')
```

```python
lamprakis_hub = np.loadtxt('lamprakis_hub.csv', delimiter=',', skiprows=1)
lamprakis_shroud = np.loadtxt('lamprakis_shroud.csv', delimiter=',', skiprows=1)

plt.figure()
plt.plot(blade.streamwise_normalized_coord[:,0], -180/np.pi*blade.blade_metal_angle[:,0], label='Hub')
plt.plot(lamprakis_hub[:,0], lamprakis_hub[:,1], 'C0o', mfc='none')
plt.plot(blade.streamwise_normalized_coord[:,-1], -180/np.pi*blade.blade_metal_angle[:,-1], label='Shroud')
plt.plot(lamprakis_shroud[:,0], lamprakis_shroud[:,1], 'C1o', mfc='none')

plt.xlabel(r'$s \ \mathrm{[-]}$')
plt.ylabel(r'$\kappa \ \mathrm{[deg]}$')
plt.legend()
plt.grid(alpha=grid_opacity)
```
