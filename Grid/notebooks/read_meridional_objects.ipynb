### Code to demonstrate how to read and use the data contained in the pickle objects that contains the meridional processed data

begin loading the files relevant to inlet, blade, and outlet blocks

```python
import pickle
import sys
sys.path.append('../../')
import Grid

filename = '../../Grid/debug/data/meta/iris_blade_30_10'
with open(filename, "rb") as file:
    blade_block = pickle.load(file)

filename = '../../Grid/debug/data/meta/iris_inlet_15_10'
with open(filename, "rb") as file:
    inlet_block = pickle.load(file)
    
filename = '../../Grid/debug/data/meta/iris_outlet_15_10'
with open(filename, "rb") as file:
    outlet_block = pickle.load(file)
```

the data are already in the 2D form, show for example the pressure field in the 3 blocks

```python
inlet_block.contour_plot(field='rho')
blade_block.contour_plot(field='rho')
outlet_block.contour_plot(field='rho')
```
